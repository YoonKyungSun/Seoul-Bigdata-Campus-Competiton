```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 서울시 주유소, 행정동 코드 파일

```python
charge = pd.read_csv("seoul_oil_sinsa.csv", encoding="cp949")

charge = charge.drop(["Unnamed: 0", "연번"], axis=1)

charge
```

```python
dong_code = pd.DataFrame(pd.read_csv("C:/R_Program_Files/R_Working_Directory/PSAT/charging_station/living_population/행정동코드_수정.csv"))

dong_code = dong_code.drop(["시도명", "시군구명"], axis=1)

dong_code.columns = ["행정동코드", "행정동명"]

dong_code
```

### 행정동별 주유소 개수

```python
charge_num = charge.groupby("행정동", as_index=False).count().iloc[:,0:2]

charge_num.columns = ["행정동", "주유소"]

charge_num
```

```python
charge_num = pd.merge(dong_code, charge_num, left_on="행정동명", right_on="행정동", how="left")

charge_num = charge_num.drop("행정동", axis=1)

charge_num["주유소"] = charge_num["주유소"].fillna(0)

charge_num = charge_num.sort_values("행정동명")

charge_num
```

### 데이터셋에 주유소 개수 추가

```python
final_charge = pd.merge(final, charge_num, left_on="행정동", right_on="행정동명", how="left")

final_charge
```
