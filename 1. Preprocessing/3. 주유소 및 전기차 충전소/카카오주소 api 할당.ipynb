```python
import requests
import json
import pandas as pd
import numpy as np
```

### 도로명주소 지번주소로 변환

```python
def doro_jibun(df):
    data = df.copy()
    data['지번주소'] = 0
    data['행정동'] = 0 
    data['경도'] = 0 
    data['위도'] = 0 
    for i in range(len(data)):
        if data.loc[i,'주소'] != '탈퇴' :
            try:
                start = data.loc[i,'주소']
                url ='https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(start)
                headers = {"Authorization": "KakaoAK " + "key값"}
                place = requests.get(url, headers = headers).json()['documents']
                temp = place[0].get('address')
                jibun = temp['address_name'] ; dong = temp['region_3depth_h_name']
                long = temp['x'] ; lat = temp['y'] 
                
                data.loc[i,['지번주소','경도','위도','행정동']] = [jibun,long,lat,dong]
                
            except:
                continue
    
    data.loc[data['지번주소']==0,['지번주소','경도','위도','행정동']] = np.nan
    return data        
```

```python
road_address = pd.read_csv('충전소별 완속 급속 충전기수 및 도로명주소.csv', encoding='cp949')

road_address
```

```python
new_road_address = doro_jibun(road_address)
new_road_address
```

### 지번주소 도로명주소로 변환

```python
def jibun_doro(df):
    data = df.copy()
    data['도로명주소'] = 0
    data['행정동'] = 0 
    data['경도'] = 0 
    data['위도'] = 0 
    for i in range(len(data)):
        if data.loc[i,'지번주소'] != '탈퇴' :
            try:
                start = data.loc[i,'지번주소']
                url ='https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(start)
                headers = {"Authorization": "KakaoAK " + "key값"}
                place = requests.get(url, headers = headers).json()['documents']
                temp = place[0].get('address')
                temp2 = place[0].get('road_address')
                doro = temp2['address_name'] ; dong = temp['region_3depth_h_name']
                long = temp['x'] ; lat = temp['y'] 
                
                data.loc[i,['도로명주소','경도','위도','행정동']] = [doro,long,lat,dong]

            except:
                continue
    
    data.loc[data['도로명주소']==0,['도로명주소','경도','위도','행정동']] = np.nan
    return data        
```
